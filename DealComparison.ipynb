```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
orders = pd.read_csv('olist_project/olist_orders_dataset.csv')
products = pd.read_csv('olist_project/olist_products_dataset.csv')
order_items = pd.read_csv('olist_project/olist_order_items_dataset.csv')
```

```python
orders_merged = orders.merge(order_items, on='order_id', how='inner') \
                                 .merge(products, on='product_id', how='inner')
```

```python
orders_merged.head()
```

```python
black_friday_orders = orders_merged[
    (
        (orders_merged['order_purchase_timestamp'] >= '2016-11-11') & 
        (orders_merged['order_purchase_timestamp'] <= '2016-11-25')
    ) |
    (
        (orders_merged['order_purchase_timestamp'] >= '2017-11-10') & 
        (orders_merged['order_purchase_timestamp'] <= '2017-11-24')
    ) |
    (
        (orders_merged['order_purchase_timestamp'] >= '2018-11-09') & 
        (orders_merged['order_purchase_timestamp'] <= '2018-11-23')
    )
]
```

```python
christmas_orders = orders_merged[
    (
        (orders_merged['order_purchase_timestamp'] >= '2016-12-11') & 
        (orders_merged['order_purchase_timestamp'] <= '2016-12-25')
    ) |
    (
        (orders_merged['order_purchase_timestamp'] >= '2017-12-11') & 
        (orders_merged['order_purchase_timestamp'] <= '2017-12-25')
    ) |
    (
        (orders_merged['order_purchase_timestamp'] >= '2018-12-11') & 
        (orders_merged['order_purchase_timestamp'] <= '2018-12-25')
    )
]
```

```python
# Group by product category and calculate the average price
black_friday_avg_price = black_friday_orders.groupby('product_category_name')['price'].mean().reset_index()
black_friday_avg_price
```

```python
# Group by product category and calculate the average price
christmas_avg_price = christmas_orders.groupby('product_category_name')['price'].mean().reset_index()
christmas_avg_price
```

```python
# Merge the Black Friday and Christmas average price DataFrames on the 'product_category' column
comparison_df = pd.merge(black_friday_avg_price, christmas_avg_price, on='product_category_name', suffixes=('_black_friday', '_christmas'))
comparison_df
```

```python
# Add a new column based on the comparison
comparison_df['better_deal'] = np.where(comparison_df['price_black_friday'] < comparison_df['price_christmas'], 'Black Friday has a better deal', 'Christmas has a better deal')

# Display the updated comparison DataFrame
comparison_df
```

```python
# Calculate the count of better deals for each occasion
better_deal_counts = comparison_df['better_deal'].value_counts()

# Create a pie chart to visualize the proportion
plt.figure(figsize=(8, 6))
plt.pie(better_deal_counts, labels=better_deal_counts.index, autopct='%1.1f%%', colors=['lightblue', 'lightgreen'])
plt.title('Proportion of Better Deals: Christmas vs. Black Friday')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

```python
# Set the size of the plot
plt.figure(figsize=(10, 6))

# Plot the average prices for Black Friday and Christmas side by side
plt.bar(comparison_df['product_category_name'], comparison_df['price_black_friday'], color='blue', alpha=0.5, label='Black Friday')
plt.bar(comparison_df['product_category_name'], comparison_df['price_christmas'], color='red', alpha=0.5, label='Christmas')

# Add labels and title
plt.xlabel('Product Category')
plt.ylabel('Average Price')
plt.title('Comparison of Average Prices between Black Friday and Christmas by Product Category')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
